```python
import csv

def read_csv_files(file_names):
    all_texts = []
    try:
        for file_name in file_names:
            with open(file_name, 'r') as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    text = row.get('text', '')  # Safely get the 'text' field
                    all_texts.append(text)
    except FileNotFoundError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")
    
    return all_texts

def write_to_txt_file(all_texts, output_file):
    try:
        with open(output_file, 'w') as outfile:
            for text in all_texts:
                outfile.write(text + '\n')
        print(f"Text written to {output_file}")
    except Exception as e:
        print(f"An error occurred while writing to the file: {e}")

csv_file_names = ['C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\CSV1.csv', 'C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\CSV2.csv', 'C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\CSV3.csv', 'C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\CSV4.csv']
all_texts = read_csv_files(csv_file_names)
write_to_txt_file(all_texts, 'C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\Assigment2.txt')
```

```python
import collections
import csv

# Define the input text file and output CSV file with double backslashes
input_file = 'C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\New folder\\your_text.txt'
output_csv_file = 'C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\New folder\\top_30_words.csv'

# Read the text from the input file
with open(input_file, 'r') as file:
    text = file.read()

# Split the text into words and convert them to lowercase
words = text.lower().split()

# Use collections.Counter to count word occurrences
word_counts = collections.Counter(words)

# Get the top 30 most common words
top_30_words = word_counts.most_common(30)

# Write the top 30 words and their counts to a CSV file
with open(output_csv_file, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Word', 'Count'])  # Write header
    for word, count in top_30_words:
        csv_writer.writerow([word, count])

print("Top 30 common words and their counts have been stored in", output_csv_file)
```

```python
#Q1T3.2
from transformers import AutoTokenizer
import re
import operator

def count_unique_tokens(file_path):
    tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
    with open(file_path, 'r') as file:
        text = file.read()
    cleaned_text = re.sub(r'\W+', ' ', text).lower()
    tokens = tokenizer.tokenize(cleaned_text)
    filtered_tokens = [token.lower() for token in tokens if token.isalnum()]
    token_counts = {}
    for token in filtered_tokens:
        if token in token_counts:
            token_counts[token] += 1
        else:
            token_counts[token] = 1

    top_30_words = sorted(token_counts.items(), key=operator.itemgetter(1), reverse=True)[:30]

    return top_30_words

file_path = "C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\Assigment2.txt"
result = count_unique_tokens(file_path)

print("Top 30 unique tokens:")
for word, count in result:
    print(f"{word}: {count}")
```

```python
import spacy
from transformers import pipeline
from collections import Counter

def extract_entities_spacy(model_name, file_path):
    nlp = spacy.load(model_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    doc = nlp(text)
    diseases = [ent.text for ent in doc.ents if ent.label_ == 'DIS']
    drugs = [ent.text for ent in doc.ents if ent.label_ == 'DRU']
    return diseases, drugs

def extract_entities_biobert(file_path):
    ner = pipeline('ner', model='dmis-lab/biobert-v1.1-adequacy', tokenizer='dmis-lab/biobert-v1.1-adequacy')
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    doc = ner(text)
    diseases = [ent['word'] for ent in doc if ent['entity'] == 'DISEASE']
    drugs = [ent['word'] for ent in doc if ent['entity'] == 'CHEMICAL']
    return diseases, drugs

def compare_results(spacy_results, biobert_results):
    total_spacy_entities = len(spacy_results[0]) + len(spacy_results[1])
    total_biobert_entities = len(biobert_results[0]) + len(biobert_results[1])

    print("Total entities detected by spaCy:", total_spacy_entities)
    print("Total entities detected by BioBERT:", total_biobert_entities)

    common_diseases = list(set(spacy_results[0]) & set(biobert_results[0]))
    common_drugs = list(set(spacy_results[1]) & set(biobert_results[1]))

    print("\nCommon diseases detected by both models:", common_diseases)
    print("Common drugs detected by both models:", common_drugs)

    print("\nMost common diseases detected by spaCy:", most_common(spacy_results[0]))
    print("Most common diseases detected by BioBERT:", most_common(biobert_results[0]))
    print("Most common drugs detected by spaCy:", most_common(spacy_results[1]))
    print("Most common drugs detected by BioBERT:", most_common(biobert_results[1]))

def most_common(lst):
    return Counter(lst).most_common()

file_path = 'Assigment2.txt'  # Replace with the path to your text file

spacy_diseases, spacy_drugs = extract_entities_spacy('en_ner_bc5cdr_md', file_path)
biobert_diseases, biobert_drugs = extract_entities_biobert(file_path)

compare_results((spacy_diseases, spacy_drugs), (biobert_diseases, biobert_drugs))
```

```python
import time
from PIL import Image

def generate_chapter1out():
    # Get the current time in seconds since the epoch (January 1, 1970).
    current_time = int(time.time())

    # Generate a number based on the current time.
    generated_number = (current_time % 100) + 50

    # Check if the generated number is even and add 10 if true.
    if generated_number % 2 == 0:
        generated_number += 10

    # Open the original image 'Chapter1.png'.
    original_image = Image.open('C:\\Users\\User\\OneDrive - Tokio Marine\\Desktop\\New folder (3)\\chapter1.jpg')

    # Create a new blank image of the same size as the original.
    new_image = Image.new('RGB', original_image.size)

    # Iterate through each pixel of the original image.
    for i in range(original_image.width):
        for j in range(original_image.height):
            r, g, b = original_image.getpixel((i, j))

            # Apply the generated_number offset to each color channel, ensuring values don't exceed 255.
            new_r = min(255, r + generated_number)
            new_g = min(255, g + generated_number)
            new_b = min(255, b + generated_number)

            # Set the modified pixel values in the new image.
            new_image.putpixel((i, j), (new_r, new_g, new_b))

    # Save the modified image as 'chapter1out.png'.
    new_image.save('chapter1out.png')

    # Calculate the sum of the red channel values in the modified image.
    red_sum = sum(new_image.getpixel((i, j))[0] for i in range(new_image.width) for j in range(new_image.height))

    return red_sum

# Call the function to generate the modified image and get the sum of red channel values.
result = generate_chapter1out()

# Print the result.
print("Sum of Red Channel Values:", result)
```
